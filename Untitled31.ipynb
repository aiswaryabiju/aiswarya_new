```python
a=df['link'].to_list()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to scrape social media profile links from a given URL
def scrape_social_media_links(url):
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        # Find div element with class "space-x-2"
        div_element = soup.find('div', class_='space-x-2')
        social_media_links = {'twitter_link': None, 'facebook_link': None, 'linkedin_link': None, 'crunchbase_link': None}
        if div_element:
            # Find anchor tags within the div
            anchors = div_element.find_all('a')
            for anchor in anchors:
                # Check if the anchor tag represents a Twitter profile
                if 'twitter.com' in anchor['href']:
                    social_media_links['twitter_link'] = anchor['href']
                # Check if the anchor tag represents a Facebook profile
                elif 'facebook.com' in anchor['href']:
                    social_media_links['facebook_link'] = anchor['href']
                # Check if the anchor tag represents a LinkedIn profile
                elif 'linkedin.com' in anchor['href']:
                    social_media_links['linkedin_link'] = anchor['href']
                # Check if the anchor tag represents a Crunchbase profile
                elif 'crunchbase.com' in anchor['href']:
                    social_media_links['crunchbase_link'] = anchor['href']
            if all(value is None for value in social_media_links.values()):
                print(f"No social media links found: {url}")
            else:
                for key, value in social_media_links.items():
                    if value is None:
                        print(f"{key.capitalize()} link not found: {url}")
                    else:
                        print(f"{key.capitalize()} link found: {url}")
        else:
            print(f"No social media links found: {url}")
        return social_media_links
    except Exception as e:
        print(f"Error scraping {url}: {e}")
        return {'twitter_link': None, 'facebook_link': None, 'linkedin_link': None, 'crunchbase_link': None}


# Scraping social media links for each URL
social_media_data = []
for url in a:
    social_media_data.append(scrape_social_media_links(url))

# Creating a DataFrame with the extracted social media profile links
df_social_media = pd.DataFrame(social_media_data)
print(df_social_media.head())
```

```python
df_social_media.to_csv("Social_media_link.csv", index=False)
```
